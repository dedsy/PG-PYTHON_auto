#### Конкатенация (Concatenation)

Конкатенация используется для объединения данных вдоль определенной оси (по вертикали или горизонтали). Метод pd.concat() используется для конкатенации объектов данных.

Пример конкатенации по вертикали:

```python
# Исходные DataFrame
df1 = pd.DataFrame({'Name': ['John', 'Alice'],
                    'Age': [25, 28]})

df2 = pd.DataFrame({'Name': ['Bob', 'Emily'],
                    'Age': [22, 30]})

# Конкатенация по вертикали
result = pd.concat([df1, df2])
```

```python
display(df1,df2,result)
```

Пример конкатенации по горизонтали:

```python
# Исходные DataFrame
df1 = pd.DataFrame({'Name': ['John', 'Alice'],
                    'Age': [25, 28]})

df2 = pd.DataFrame({'City': ['New York', 'London'],
                    'Salary': [50000, 60000]})

# Конкатенация по горизонтали
result = pd.concat([df1, df2], axis=1)
display(df1,df2,result)
```

#### Слияние (Merge)

Слияние используется для объединения наборов данных на основе значений общих столбцов. Метод pd.merge() используется для слияния данных.

Пример слияния по общему столбцу:

```python
# Исходные DataFrame
df1 = pd.DataFrame({'Name': ['John', 'Alice','Bob'],
                    'Age': [25, 28,30],
                    'City': ['New York', 'London','Kiev']})

df2 = pd.DataFrame({'Name': ['John', 'Alice'],
                    'Salary': [50000, 60000]})

# Слияние по общему столбцу "Name"
result = pd.merge(df1, df2, on='Name')
display(df1,df2,result)
```

#### Соединение (Join)

Соединение используется для объединения наборов данных на основе индексов. Метод .join() используется для соединения данных.

Пример соединения по индексу:

```python
# Исходные DataFrame
df1 = pd.DataFrame({'Age': [25, 28]},
                   index=['John', 'Alice'])

df2 = pd.DataFrame({'City': ['New York', 'London']},
                   index=['John', 'Alice'])

# Соединение по индексу
result = df1.join(df2)
display(df1,df2,result)
```

#### Агрегирование позволяет суммировать, подсчитывать статистики и выполнять другие операции над данными. Группировка позволяет разделить данные на группы в соответствии с заданными критериями и выполнять операции агрегирования для каждой группы.

```python
# Пример агрегирования суммы

df = pd.DataFrame({'Category': ['A', 'A', 'B', 'B', 'A'],
                   'Value': [10, 20, 15, 25, 30]})

# Агрегирование суммы по категориям
result = df.groupby('Category')['Value'].sum()
display(df,result)
```

```python
# Агрегирование среднего значения по категориям
result = df.groupby('Category')['Value'].mean()
result
```

```python
import seaborn as sns
```

```python
# Загрузка встроенных наборов данных в Seaborn
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
display(iris.head(), tips.head())
```

```python
# Гистограмма
sns.histplot(data=tips, x='total_bill', kde=True)
plt.xlabel('Общая сумма счета')
plt.ylabel('Частота')
plt.title('Гистограмма общей суммы счета')
plt.show()
```

```python
# Диаграмма рассеяния
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.xlabel('Длина')
plt.ylabel('Ширина')
plt.title('Диаграмма рассеяния')
plt.show()
```

```python
# Ящик с усами
sns.boxplot(data=iris, x='species', y='petal_length')
plt.xlabel('Вид')
plt.ylabel('Длина лепестка')
plt.title('Ящик с усами длины лепестка')
plt.show()
```

```python
# Настройка стилей
sns.set_style('whitegrid')  # Установка стиля
sns.histplot(data=tips, x='total_bill', kde=True)
plt.xlabel('Общая сумма счета')
plt.ylabel('Частота')
plt.title('Гистограмма общей суммы счета')
plt.show()

# Настройка цветовой палитры
sns.set_palette('pastel')  # Установка палитры цветов
sns.boxplot(data=iris, x='species', y='petal_length')
plt.xlabel('Вид')
plt.ylabel('Длина лепестка')
plt.title('Ящик с усами длины лепестка')
plt.show()
```

```python
# Тепловая карта (матрица корреляции)
corr_matrix = iris.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Тепловая карта матрицы корреляции')
plt.show()

# Facet Grid (сетка с подграфиками)
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
plt.show()
```
